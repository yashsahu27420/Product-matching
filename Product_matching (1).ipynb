```python
# importing libraries in python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# importing amazon data
dataa= pd.read_excel('https://github.com/yashsahu27420/Product-matching/blob/main/amz_com_ecommerce_sample.xlsx?raw=true')
pamz=dataa['product_name']
amz=dataa[['website','product_name','retail_price','discounted_price']]
dataa.head()
```

```python
# importing flipkart data
dataf = pd.read_excel('https://github.com/yashsahu27420/Product-matching/blob/main/flipkart_com-ecommerce_sample.xlsx?raw=true')
pflip=dataf['product_name']
flip=dataf[['website','product_name','retail_price','discounted_price']]
dataf
```

```python
# TAKING INPUT FROM USER OF PRODUCT 
prod= input("Product Name: ")
prod
```

```python
def get_matches(query,choices=2):
  amazon= process.extractOne(query, choices)
  return amazon
```

```python
#FINDING MATCH IN AMAZON
a=get_matches(prod,pamz)
z = a[2]
```

```python
#FINDING MATCH OF THE INPUT IN FLIPKART DATA
def get_matches(query,choices):
  flipkart= process.extractOne(query,choices)
  return flipkart
```

```python
#FINDING MATCH IN FLIPKART
f=get_matches(prod,pflip)
k =f[2]
```

```python
#extracting Row number
amzfin=flip.loc[k]
flipfin =amz.loc[z]
```

```python
#creating comman variable
final= (flipfin,amzfin)
```

```python
# the final output
a=list(final)
table=pd.DataFrame(a)
table
```
